<h2>Introduction to Pandas and Data Types</h2>
<p><strong>Pandas</strong> is a powerful open-source Python library designed for data manipulation and analysis. Built on top of NumPy, it provides easy-to-use data structures like Series (1D) and DataFrame (2D) to handle structured data efficiently. Whether you're cleaning messy datasets, performing statistical analysis, or preparing data for machine learning, Pandas is your go-to tool. One of its key features is its flexible handling of data types, which allows you to optimize memory usage and ensure accurate computations. Let‚Äôs dive into the types of data Pandas supports and how they work!</p>
<img src="./assets/pandas.png" class="image-placeholder">

<h3>Numeric Data Types</h3>
<p>Pandas (and NumPy, which it builds on) offers a variety of numeric data types for integers and floating-point numbers:</p>
<ul>
    <li><strong>Integers:</strong>
        <ul>
            <li><code>int8, int16, int32, int64</code>: Signed integers (can be negative or positive) with sizes of 8, 16, 32, or 64 bits.</li>
            <li><code>uint8, uint16, uint32, uint64</code>: Unsigned integers (only positive) with the same bit sizes.</li>
        </ul>
    </li>
    <li><strong>Floating-Point Numbers:</strong>
        <ul>
            <li><code>float16, float32, float64</code>: Floating-point numbers with 16, 32, or 64 bits of precision. <code>float64</code> is also called "double" and is the default in Pandas.</li>
        </ul>
    </li>
</ul>
<p><strong>Example:</strong></p>
<pre>
import pandas as pd
df = pd.DataFrame({'A': [1, 2, 3], 'B': [1.5, 2.7, 3.2]})
df['A'] = df['A'].astype('int32')  # Convert to 32-bit integer
df['B'] = df['B'].astype('float32')  # Convert to 32-bit float
print(df.dtypes)
# Output:
# A     int32
# B    float32
</pre>

<h3>String Data Types</h3>
<p>Strings are handled differently in NumPy and Pandas:</p>
<ul>
    <li><strong>NumPy:</strong>
        <ul>
            <li><code>dtype='U'</code>: Unicode strings (variable length).</li>
            <li><code>dtype='S'</code>: Byte strings (fixed length).</li>
        </ul>
    </li>
    <li><strong>Pandas:</strong>
        <ul>
            <li><code>object</code>: Default type for strings or mixed data in Pandas.</li>
            <li><code>string</code>: A newer Pandas-specific type (introduced in version 1.0.0) for pure string data, more efficient than <code>object</code>.</li>
        </ul>
    </li>
</ul>
<p><strong>Example:</strong></p>
<pre>
df = pd.DataFrame({'Names': ['Ali', 'Sara', 'Omid']})
print(df.dtypes)  # Names: object
df['Names'] = df['Names'].astype('string')  # Convert to string type
print(df.dtypes)  # Names: string
</pre>

<h3>Boolean Data Type</h3>
<p>Both Pandas and NumPy support a boolean type:</p>
<ul>
    <li><code>bool</code>: Represents <code>True</code> or <code>False</code> values.</li>
</ul>
<p><strong>Example:</strong></p>
<pre>
df = pd.DataFrame({'Active': [True, False, True]})
print(df.dtypes)  # Active: bool
</pre>

<h3>DateTime Data Types</h3>
<p>Handling dates and times is essential for time-series analysis:</p>
<ul>
    <li><strong>NumPy:</strong>
        <ul>
            <li><code>datetime64</code>: General type for dates and times.</li>
        </ul>
    </li>
    <li><strong>Pandas:</strong>
        <ul>
            <li><code>datetime64[ns]</code>: High-precision datetime with nanosecond resolution.</li>
            <li><code>timedelta64[ns]</code>: For differences between times (e.g., durations).</li>
        </ul>
    </li>
</ul>
<p><strong>Example:</strong></p>
<pre>
df = pd.DataFrame({'Dates': ['2023-01-01', '2023-01-02']})
df['Dates'] = pd.to_datetime(df['Dates'])  # Convert to datetime64[ns]
print(df.dtypes)  # Dates: datetime64[ns]
df['Diff'] = df['Dates'] - pd.to_datetime('2023-01-01')
print(df.dtypes)  # Diff: timedelta64[ns]
</pre>

<h3>Categorical Data Type</h3>
<p>Pandas provides a special type for categorical data, which is great for repeated values:</p>
<ul>
    <li><code>category</code>: Stores unique values efficiently and is ideal for columns with limited, repeated entries (e.g., gender, days of the week).</li>
</ul>
<p><strong>Example:</strong></p>
<pre>
df = pd.DataFrame({'Day': ['Mon', 'Tue', 'Mon', 'Wed']})
df['Day'] = df['Day'].astype('category')
print(df.dtypes)  # Day: category
print(df['Day'].cat.categories)  # Index(['Mon', 'Tue', 'Wed'], dtype='object')
</pre>

<h3>Special Data Types</h3>
<p>Some additional types for specific use cases:</p>
<ul>
    <li><strong>NumPy:</strong>
        <ul>
            <li><code>complex</code>: For complex numbers (e.g., 3 + 4j).</li>
        </ul>
    </li>
    <li><strong>Pandas:</strong>
        <ul>
            <li><code>Int64, Float64</code>: Nullable integer and float types that can handle <code>NaN</code> without converting to float.</li>
            <li><code>sparse</code>: For datasets with lots of zeros, saving memory.</li>
        </ul>
    </li>
</ul>
<p><strong>Example:</strong></p>
<pre>
df = pd.DataFrame({'A': [1, None, 3]})
df['A'] = df['A'].astype('Int64')  # Nullable integer
print(df.dtypes)  # A: Int64
print(df)  # Can keep NaN without changing type
</pre>

<h3>The <code>dtype</code> Parameter</h3>
<p>The <code>dtype</code> parameter in Pandas (and NumPy) lets you specify the data type when creating or converting data. It‚Äôs super useful for controlling memory usage and ensuring your data behaves as expected. You can pass it to functions like <code>pd.DataFrame()</code>, <code>pd.Series()</code>, or <code>astype()</code>.</p>
<p><strong>Example:</strong></p>
<pre>
# Creating a DataFrame with specific dtypes
df = pd.DataFrame({'A': [1, 2, 3], 'B': ['x', 'y', 'z']}, dtype={'A': 'int8', 'B': 'string'})
print(df.dtypes)  # A: int8, B: string

# Converting with astype
df['A'] = df['A'].astype('float32')
print(df.dtypes)  # A: float32, B: string
</pre>

<h2>Loading and Exploring Data</h2>
<p>Welcome to this section on <strong>Exploratory Data Analysis (EDA)</strong>! Before building complex models, we need to understand our data. This section will guide you through the initial steps of working with data in Python, exploring it, and preparing it for machine learning. Let‚Äôs dive in!</p>

<h3>Loading Data into Python</h3>
<p>Datasets can come in various formats (CSV, Excel, JSON, etc.), and Python offers tools to load them:</p>
<ul>
    <li><strong>CSV Files:</strong> Use <code>pandas.read_csv()</code> to load comma-separated value files.</li>
    <li><strong>Excel Files:</strong> Use <code>pandas.read_excel()</code> for spreadsheets.</li>
    <li><strong>Built-in Datasets:</strong> Libraries like <code>seaborn</code> provide sample datasets (e.g., "tips", "iris") via <code>sns.load_dataset()</code>.</li>
</ul>

<h3>Basic Exploration Methods</h3>
<p>Once loaded, use these Pandas methods to get a quick overview:</p>
<ul>
    <li><strong>.head(n):</strong> Displays the first <code>n</code> rows of the dataset (default is 5). Great for a quick peek at the data.</li>
    <li><strong>.info():</strong> Shows a summary of the dataset‚Äîcolumn names, data types, and missing values. Perfect for understanding structure.</li>
    <li><strong>.value_counts():</strong> Counts unique values in a column. Useful for categorical data (e.g., how many times each category appears).</li>
    <li><strong>.describe():</strong> Provides statistical summaries (mean, min, max, etc.) for numerical columns. A fast way to see data ranges and central tendencies.</li>
</ul>

<section id="section1">
    <h2>Introduction to Test Set Creation</h2>
    <p>Before exploring further, it‚Äôs critical to set aside a test set (and sometimes a validation set) to evaluate your model later. This prevents "data snooping" bias‚Äîwhere you unintentionally overfit to the entire dataset.</p>
    <img src="https://algotrading101.com/learn/wp-content/uploads/2020/06/training-validation-test-data-set.png" class="image-placeholder">
</section>

<section id="section2">
    <h2>Basic Train-Test Split</h2>
    <p>Use <code>train_test_split</code> from <code>sklearn.model_selection</code>:</p>
    <ul>
        <li>Splits data into training (e.g., 80%) and test (e.g., 20%) sets.</li>
        <li><code>random_state</code> ensures reproducibility.</li>
    </ul>
</section>

<section id="section3">
    <h2>Train-Validation-Test Split</h2>
    <p>For tuning models, split data into three parts:</p>
    <ul>
        <li>Training (e.g., 60%): For model learning.</li>
        <li>Validation (e.g., 20%): For tuning hyperparameters.</li>
        <li>Test (e.g., 20%): For final evaluation.</li>
    </ul>
    <p>Can be done with multiple <code>train_test_split</code> calls.</p>
</section>

<section id="section4">
    <h2>StratifiedShuffleSplit</h2>
    <p><strong>StratifiedShuffleSplit</strong> is a powerful tool from <code>sklearn.model_selection</code> that combines stratified sampling with shuffling to create multiple train-test splits. Unlike a single <code>train_test_split</code>, it generates several splits while ensuring that the proportion of categories (e.g., classes or groups) remains consistent across all splits. This is especially useful for cross-validation, evaluating model stability, or when you need repeated stratified samples from an imbalanced dataset.</p>
    <p><strong>Key Points:</strong></p>
    <ul>
        <li><strong>When to Use:</strong> For cross-validation with imbalanced data or when you want multiple stratified splits.</li>
        <li><strong>How It Works:</strong> It shuffles the data, then splits it into training and test sets multiple times, maintaining category proportions based on a specified column.</li>
        <li><strong>Parameters:</strong>
            <ul>
                <li><code>n_splits</code>: Number of re-shuffling & splitting iterations.</li>
                <li><code>test_size</code>: Proportion of data for the test set.</li>
                <li><code>train_size</code>: Proportion for the training set (optional).</li>
                <li><code>random_state</code>: For reproducibility.</li>
            </ul>
        </li>
    </ul>
</section>

```python
from sklearn.model_selection import train_test_split

# Basic Train-Test Split
train_set, test_set = train_test_split(combined_df, test_size=0.2, random_state=42)
print("Training set size:", len(train_set))
print("Test set size:", len(test_set))

# Train-Validation-Test Split
train_val, test_set = train_test_split(combined_df, test_size=0.2, random_state=42)
train_set, val_set = train_test_split(train_val, test_size=0.25, random_state=42)  # 0.25 of 80% = 20% of original
print("\nTrain set size:", len(train_set))
print("Validation set size:", len(val_set))
print("Test set size:", len(test_set))
```

<section id="section5">
<h2>Introduction to Categorical Variables</h2>
<p>Categorical variables are a fundamental data type in machine learning and statistical analysis. These variables represent qualitative data that can be divided into distinct categories or groups. Understanding and properly handling categorical variables is crucial for effective data preprocessing and model performance.</p>
<p>Categorical variables, also known as nominal variables, are variables that can take on one of a limited number of possible values or categories. Unlike numerical variables, categorical variables don't have a natural numerical order or magnitude.</p>
<p><strong>Examples of categorical variables include:</strong></p>
<ul>
    <li>Color (red, blue, green)</li>
    <li>Gender (male, female, non-binary)</li>
    <li>Country (USA, Canada, Japan)</li>
    <li>Product type (electronics, clothing, food)</li>
</ul>
</section>

<section id="section6">
<h2>One-Hot Encoding</h2>
<p><strong>One-Hot Encoding</strong> is a popular and widely used technique for handling categorical variables in machine learning. It's particularly useful when dealing with nominal categorical variables, where there's no inherent order among the categories. This method creates new binary columns for each category, effectively representing the presence or absence of a category for each observation.</p>
<p><strong>Understanding One-Hot Encoding:</strong> One-Hot Encoding works by creating a new binary column for each unique category in the original categorical variable. Each observation will have a '1' in the column corresponding to its category and '0' in all other columns.</p>
<p class="highlight">üîë <strong>Key Concept:</strong> One-Hot Encoding transforms a single categorical variable into multiple binary variables, each representing a unique category.</p>
</section>

```python
import pandas as pd
from sklearn.model_selection import train_test_split
from sklearn.linear_model import LinearRegression
from sklearn.preprocessing import OneHotEncoder
from sklearn.metrics import mean_squared_error, r2_score
import scipy.sparse
import numpy as np

X = train_set.drop(columns=['price', 'date'])
y = train_set['price']

categorical_cols = X.select_dtypes(include='category').columns.tolist()
numerical_cols = X.select_dtypes(exclude='category').columns.tolist()

encoder = OneHotEncoder(handle_unknown='ignore', sparse_output=True)
X_encoded = encoder.fit_transform(X[categorical_cols])

X_numerical = X[numerical_cols].values

X_all = scipy.sparse.hstack((X_encoded, X_numerical), format='csr')

model = LinearRegression()
model.fit(X_all, y)

y_pred = model.predict(X_all)
mse = mean_squared_error(y, y_pred)

print(f"Mean Squared Error: {mse:.2f}")
```

```python
import pandas as pd
from sklearn.model_selection import train_test_split
from sklearn.linear_model import LinearRegression
from sklearn.preprocessing import OneHotEncoder
from sklearn.compose import ColumnTransformer
from sklearn.pipeline import Pipeline
from sklearn.metrics import mean_squared_error
from scipy.sparse import hstack
import numpy as np

X = train_set.drop(columns=['price', 'date'])
y = train_set['price']

categorical_cols = X.select_dtypes(include='category').columns.tolist()
numerical_cols = X.select_dtypes(exclude='category').columns.tolist()

preprocessor = ColumnTransformer([
    ('cat', OneHotEncoder(handle_unknown='ignore', sparse_output=True), categorical_cols),
    ('num', 'passthrough', numerical_cols)
], sparse_threshold=1.0)

pipeline = Pipeline([
    ('preprocessing', preprocessor),
    ('regressor', LinearRegression())
])

pipeline.fit(X, y)

y_pred = pipeline.predict(X)
mse = mean_squared_error(y, y_pred)

print(f"Mean Squared Error: {mse:.2f}")
```

<section id="introduction">
<h2>Introduction to Feature Scaling</h2>
<p>Feature scaling is a crucial preprocessing step in machine learning to ensure that features with different scales or units contribute equally to the model‚Äôs performance. Many algorithms, such as those relying on gradient descent or distance metrics, are sensitive to the scale of input features. Scaling transforms features to a common range or distribution, improving model convergence and accuracy.</p>
<p class="highlight">üîë <strong>Key Concept:</strong> Scaling prevents features with larger ranges from dominating the model, enabling fair comparisons and better optimization.</p>
</section>

<section id="standard-scaler">
<h2>StandardScaler</h2>
<p><strong>Definition:</strong> StandardScaler standardizes features by removing the mean and scaling to unit variance, transforming data to have a mean of 0 and a standard deviation of 1. It assumes the data is approximately normally distributed.</p>
<p><strong>How it Works:</strong> For each feature \( x \), it applies the transformation:</p>
<div class="math-block">
    z = \frac{x - \mu}{\sigma}
</div>
<p>where \( \mu \) is the mean and \( \sigma \) is the standard deviation of the feature.</p>
<p><strong>Advantages:</strong></p>
<ul>
    <li>Effective for algorithms assuming normally distributed data, such as linear regression or logistic regression.</li>
    <li>Robust to moderate variations in data scale.</li>
</ul>
<p><strong>Disadvantages:</strong></p>
<ul>
    <li>Sensitive to outliers, as the mean and standard deviation are influenced by extreme values.</li>
    <li>Assumes data is Gaussian-like, which may not always be true.</li>
</ul>
<p><strong>Recommended Models:</strong></p>
<ul>
    <li>Linear Regression, Logistic Regression, Support Vector Machines (SVM), Principal Component Analysis (PCA).</li>
    <li>Algorithms using gradient descent or distance metrics (e.g., k-Nearest Neighbors).</li>
</ul>
<p><strong>Use Cases:</strong> When features have different units (e.g., height in cm, weight in kg) and the algorithm benefits from standardized inputs.</p>
</section>

<section id="minmax-scaler">
<h2>MinMaxScaler</h2>
<p><strong>Definition:</strong> MinMaxScaler transforms features to a fixed range, typically [0, 1], by scaling based on the minimum and maximum values of the feature.</p>
<p><strong>How it Works:</strong> For each feature \( x \), it applies:</p>
<div class="math-block">
    x' = \frac{x - x_{min}}{x_{max} - x_{min}}
</div>
<p><strong>Advantages:</strong></p>
<ul>
    <li>Preserves the original distribution shape and relationships between values.</li>
    <li>Ideal for algorithms requiring bounded input ranges, such as neural networks.</li>
</ul>
<p><strong>Disadvantages:</strong></p>
<ul>
    <li>Highly sensitive to outliers, as \( x_{min} \) and \( x_{max} \) are directly affected by extreme values.</li>
    <li>Not suitable for data with heavy-tailed distributions.</li>
</ul>
<p><strong>Recommended Models:</strong></p>
<ul>
    <li>Neural Networks, k-Nearest Neighbors (k-NN), Support Vector Machines (SVM).</li>
    <li>Algorithms sensitive to feature ranges.</li>
</ul>
<p><strong>Use Cases:</strong> When data needs to be constrained to a specific range or when working with neural networks that perform better with normalized inputs.</p>
</section>

<section id="comparison">
<h2>Comparison and Recommendations</h2>
<p><strong>Differences:</strong></p>
<ul>
    <li><strong>StandardScaler</strong> centers data to a mean of 0 and unit variance, making it ideal for Gaussian-like data but sensitive to outliers.</li>
    <li><strong>MinMaxScaler</strong> bounds data to a specific range (e.g., [0, 1]), suitable for neural networks but also sensitive to outliers.</li>
</ul>
<p><strong>Recommendations:</strong></p>
<ul>
    <li>Use <code>StandardScaler</code> for gradient-based models (e.g., Linear Regression, SVM) or PCA when data is near-Gaussian and outliers are minimal.</li>
    <li>Use <code>MinMaxScaler</code> for neural networks or algorithms requiring bounded inputs, especially when data has no significant outliers.</li>
</ul>
<p><strong>Key Considerations:</strong> Test both scalers using cross-validation to evaluate their impact on model performance. The choice depends on the data distribution, presence of outliers, and the algorithm‚Äôs requirements.</p>
</section>
