<h2 style="color: white; text-align: center; margin-bottom: 20px; font-weight: bold;">What is Web Scraping?</h2>
    <ul style="padding-left: 20px; margin-bottom: 20px; color: #d4d4d4;">
        <li><strong style="color: white;">Definition:</strong> Web scraping is the process of automatically extracting data from websites.</li>
        <li><strong style="color: white;">Purpose:</strong> It collects information like text, images, or prices for analysis or use in applications.</li>
        <li><strong style="color: white;">Tools:</strong> Uses libraries like Beautiful Soup, Requests, or Selenium to navigate and parse web pages.</li>
        <li><strong style="color: white;">Legality:</strong> Depends on the website's terms; always check for permission to avoid legal issues.</li>
    </ul>
